
# 图生图-ControlNet-Canny

## 进入 MMagic 主目录

```python
import os
os.chdir('/content/drive/MyDrive/OpenMMlab/mmagic')
```

## 导入工具包

```python
import cv2
import numpy as np
import mmcv
from mmengine import Config
from PIL import Image

from mmagic.registry import MODELS
from mmagic.utils import register_all_modules

register_all_modules()
```

## 载入ControlNet模型

```python
cfg = Config.fromfile('configs/controlnet/controlnet-canny.py')
controlnet = MODELS.build(cfg.model).cuda()
```

## 下载图像，做Canny边缘检测

```python
%mkdir data
```

```python
import cv2
import numpy as np
import requests
import io


# 下载图片
url = 'https://bkimg.cdn.bcebos.com/pic/e824b899a9014c08f148680a007b02087bf4f42a'
response = requests.get(url)
image = np.asarray(bytearray(response.content), dtype="uint8")
image = cv2.imdecode(image, cv2.IMREAD_GRAYSCALE)

# 保存原始图片
cv2.imwrite('data/image.jpg', image)

# 使用Canny边缘检测
edges = cv2.Canny(image, threshold1=30, threshold2=100)

# 保存图像
cv2.imwrite('data/edges.jpg', edges)
```

## 对比边缘检测前后图

```python
import matplotlib.pyplot as plt

# 读取原图
original = cv2.imread('data/image.jpg', cv2.IMREAD_COLOR)

# 读取处理后的图像
edges = cv2.imread('data/edges.jpg', cv2.IMREAD_COLOR)

# OpenCV使用BGR模式，而matplotlib使用RGB模式，所以我们需要转换颜色模式
original = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)
edges = cv2.cvtColor(edges, cv2.COLOR_BGR2RGB)

# 创建子图来展示对比
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# 展示原图
axs[0].imshow(original, cmap='gray')
axs[0].set_title('Original Image')

# 展示处理后的图像
axs[1].imshow(edges, cmap='gray')
axs[1].set_title('Edge Image')

# 显示图像
plt.show()
```

## 输入Canny边缘图

```python
control_url = 'data/edges.jpg'
control_img = mmcv.imread(control_url)
control = cv2.Canny(control_img, 100, 200)
control = control[:, :, None]
control = np.concatenate([control] * 3, axis=2)
control = Image.fromarray(control)
```

## 咒语Prompt

```python
prompt = 'baby room'
```

## 执行预测

```python
output_dict = controlnet.infer(prompt, control=control)
samples = output_dict['samples']
for idx, sample in enumerate(samples):
    sample.save(f'sample_{idx}.png')
controls = output_dict['controls']
for idx, control in enumerate(controls):
    control.save(f'control_{idx}.png')
```

# 结果展示

```python
import matplotlib.pyplot as plt
import cv2

# 以彩色模式读取图片
original = cv2.imread('data/image.jpg', cv2.IMREAD_COLOR)
edges = cv2.imread('data/edges.jpg', cv2.IMREAD_COLOR)
sample = cv2.imread('sample_0.png', cv2.IMREAD_COLOR)

# OpenCV使用BGR模式，而matplotlib使用RGB模式，所以我们需要转换颜色模式
original = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)
edges = cv2.cvtColor(edges, cv2.COLOR_BGR2RGB)
sample = cv2.cvtColor(sample, cv2.COLOR_BGR2RGB)


# 创建子图来展示对比
fig, axs = plt.subplots(1, 3, figsize=(10, 5))

# 展示原图
axs[0].imshow(original)
axs[0].set_title('Original Image')

# 展示边缘图像
axs[1].imshow(edges)
axs[1].set_title('Edge Image')

# 展示处理后的图像
axs[2].imshow(sample)
axs[2].set_title('Sample Image')

# 显示图像
plt.suptitle('Prompt:"baby room"')
plt.show()
```
